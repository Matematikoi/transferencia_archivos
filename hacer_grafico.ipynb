# Resultados de tiempos de transferencia


En este experimento tomamos cuatro formas de transferir archivos, en particular mediante archivos, sockets, pipes y memoria compartida y comparamos el tiempo que nos lleva el llevar una información de un proceso a otro. La información que pasamos es un arreglo de enteros que llenamos con información, y le pasamos el arreglo al otro proceso por medio del método seleccionado.

Para tomar los tiempos usamos el promedio de 3 corridas y consideramos mensajes con pesos de 1kb, 10kb, 100kb, 1mb, 10mb y 100mb. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.style as style
style.use('seaborn-poster') #sets the size of the charts
style.use('ggplot')
```

```python
data = pd.read_csv('acumulado.csv')
data['tiempo'] = (data.t1+data.t2+data.t3)/3.0
data.set_index('peso',inplace = True)
data.groupby('tipo_transferencia')['tiempo'].plot(legend=True)
plt.xlabel("Peso del archivo")
plt.ylabel("Segundos")
plt.title("Comparación de velocidad entre distintos\nmetodos de transferencia entre procesos")
plt.savefig("imagen_resultado.png",dpi=300)
```
